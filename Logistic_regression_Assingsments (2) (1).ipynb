```python
df.isna().sum()
```

```python
df_test['Fare']=df_test['Fare'].fillna(df_test['Fare'].median())
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df_test['Age']=df_test['Age'].fillna(df_test['Age'].median())
```

```python
df.drop(columns=['Cabin'],inplace=True)
```

```python
df_test.drop(columns=['Cabin'],inplace=True)
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df_test['Embarked']=df_test['Embarked'].fillna(df_test['Embarked'].mode()[0])
```

```python
df.drop(columns=['PassengerId','Name','Ticket'],inplace=True)
```

```python
df_test.drop(columns=['PassengerId','Name','Ticket'],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
df['Embarked']=le.fit_transform(df['Embarked'])
df
```

```python
df_test['Sex']=le.fit_transform(df_test['Sex'])
df_test['Embarked']=le.fit_transform(df_test['Embarked'])
```

```python
sns.pairplot(data=df,hue='Survived')
```

```python
feature = ['Pclass',	'Sex',	'Age',	'SibSp',	'Parch',	'Fare',	'Embarked']
```

```python
x=df[feature]

y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2,random_state=42)
```

#3. Model Building:

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_val)
```

```python
y_pred
```

#4.Model Evaluation:

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Assuming 'x1' and 'y1' are your test features and labels, respectively
# Make predictions on the test data
y_pred_test = lr.predict(x_val)

# Calculate evaluation metrics on the test data
accuracy_test = accuracy_score(y_val, y_pred_test)
precision_test = precision_score(y_val, y_pred_test)
recall_test = recall_score(y_val, y_pred_test)
f1_test = f1_score(y_val, y_pred_test)
roc_auc_test = roc_auc_score(y_val, y_pred_test)

# Print the evaluation metrics
print("Evaluation Metrics on Test Data:")
print(f'Accuracy: {accuracy_test:.4f}')
print(f'Precision: {precision_test:.4f}')
print(f'Recall: {recall_test:.4f}')
print(f'F1-score: {f1_test:.4f}')
print(f'ROC-AUC: {roc_auc_test:.4f}')
```

```python
from sklearn.metrics import roc_curve, auc


# Get the probability of the positive class (survival)
y_prob = lr.predict_proba(x_val)[:, 1]

# Calculate the false positive rate, true positive rate, and thresholds
fpr, tpr, thresholds = roc_curve(y_val, y_prob)

# Calculate the AUC
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

#b. Discuss the significance of features in predicting the target variable (survival probability in this case).
Ans:

Based on the coefficients we examined, we can discuss the significance of each feature in predicting the target variable (survival probability).

The magnitude of the absolute value of the coefficient indicates the strength of the relationship between the feature and the log-odds of survival. A larger absolute value suggests a stronger influence. The sign of the coefficient indicates the direction of the relationship (positive for increasing log-odds of survival, negative for decreasing).

Here's a discussion of the significance of each feature:

Sex: With the largest absolute coefficient (-2.60), 'Sex' is the most significant feature in predicting survival probability. Being male has a strong negative impact on the likelihood of survival.
Pclass: 'Pclass' is the second most significant feature with an absolute coefficient of 0.93. Being in a higher class (lower Pclass value) significantly increases the log-odds of survival.
Age: 'Age' has a relatively small absolute coefficient (0.03), suggesting a weaker influence on survival probability compared to 'Sex' and 'Pclass'. While older age slightly decreases the log-odds of survival, its impact is not as substantial.
SibSp and Parch: Both 'SibSp' and 'Parch' have relatively small negative coefficients (-0.29 and -0.11 respectively), indicating that having more siblings/spouses or parents/children on board has a slight negative impact on the log-odds of survival. Their significance is less pronounced than 'Sex' and 'Pclass'.
Embarked: 'Embarked' has a relatively small negative coefficient (-0.21), suggesting a minor influence on survival probability depending on the embarkation point.
Fare: 'Fare' has the smallest absolute coefficient (0.0025), suggesting it is the least significant feature among those considered in predicting survival probability. While a higher fare slightly increases the log-odds of survival, its impact is minimal.
In summary, 'Sex' and 'Pclass' are the most significant features in this logistic regression model for predicting survival probability on the Titanic, followed by 'Age'. The other features ('SibSp', 'Parch', 'Embarked', and 'Fare') have less significant impacts.
